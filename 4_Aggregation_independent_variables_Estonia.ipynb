```python
import geopandas as gpd
import pandas as pd
```

```python
# reading files
# est
est_path = r'data\gwr_file\est_gwr.gpkg'
est = gpd.read_file(est_path, driver='GPKG')

est['County'] = est['to_fi_from'].str.replace('maakond', 'county')

est.plot()
est.head(2)
```

```python
# UNEMPLOYMENT RATE
un_path = r'datasets\Estonia\unemploymentRate_EST_2019.csv'
est_unem = pd.read_csv(un_path, encoding='utf-8', skiprows=2)

est_unem = est_unem.rename(columns={'2019':'unem_%19'})[['County', 'unem_%19']]

# est_unem

est = est.merge(est_unem, on='County')
est.head()
```

```python
# GDP PER CAPITA
gdp_path = r'datasets\Estonia\GDP_perCapita_EST_2019.csv'
est_gdp = pd.read_csv(gdp_path, encoding='utf-8', skiprows=2)

est_gdp = est_gdp.rename(columns={'GDP per capita, euros':'gdp_eur'})[['County', 'gdp_eur']]

# est_gdp.head()

est = est.merge(est_gdp, on='County')
est.head()
```

```python
# POPULATION OVER 65%
pop_path = r'datasets\Estonia\age_over65_EST_2019.csv'
est_pop = pd.read_csv(pop_path, encoding='utf-8', skiprows=2)

est_pop = est_pop.rename(columns={'% over 65':'pop%65'})[['County', 'pop%65']]

# est_pop.head()

est = est.merge(est_pop, on='County')
est.head()
```

```python
# EDUCATION
edu_path = r'datasets\Estonia\education_EST_2019.csv'
est_gen = pd.read_csv(edu_path, encoding='utf-8', skiprows=2, sep=',')

est_gen = est_gen.rename(columns={'%highered':'pop%edu'})[['County', 'pop%edu']]

est_gen.head()

est = est.merge(est_gen, on='County')
est.head()
```

```python
est = est[['to_fi_from', 'MKOOD', 'County', 'rate', 'em_2019', 'unem_%19', 'gdp_eur', 'dist', 'pop%65',  'pop%edu', 'geometry']]

est = est.rename(columns={'rate': 'im%_rate', 'dist':'proxkm' })
```

```python
est.plot(column = 'im%_rate', legend=True)
est.plot(column = 'em_2019', legend=True)
est.plot(column = 'unem_%19', legend=True)
est.plot(column = 'proxkm', legend=True)
est.plot(column = 'pop%65', legend=True)
est.plot(column = 'pop%edu', legend=True)
est.head()
```

```python
from shapely.geometry import Point

est['POLYID'] = list(range(len(est)))

for row, item in est['geometry'].items():
    est.at[row,'X'] = item.centroid.x 
    est.at[row,'Y'] = item.centroid.y 

est.head()
```

```python
# SAVING FINAL FILE FOR WGR
est_gwr_finalpathg= r'data\gwr_file\final\est_gwr_final.gpkg'
est_gwr_finalpaths= r'data\gwr_file\final\est_gwr_final.shp'

est.to_file(est_gwr_finalpathg, driver='GPKG')
est.to_file(est_gwr_finalpaths)
```
